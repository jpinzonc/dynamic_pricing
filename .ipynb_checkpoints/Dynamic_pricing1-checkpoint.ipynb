```python
beef = pd.read_csv('beef_data.csv', header = 0)
beef.columns
```

```python
beef_model = smf.OLS(beef.Quantity, beef.Price).fit()
print(beef_model.summary())
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std

prstd, iv_l, iv_u = wls_prediction_std(beef_model)
x = beef.Quantity

y = beef.Price 
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x, y, 'o', label="data")
#ax.plot(x, y_true, 'b-', label="True")
ax.plot(x, beef_model.fittedvalues, 'r--.', label="OLS")
ax.plot(x, iv_u, 'r--')
ax.plot(x, iv_l, 'r--')
ax.legend(loc='best');
```

```python
import seaborn as sns

sns.regplot(y='Quantity', x='Price', data=beef)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(beef_model, fig=fig)
```

```python
fig = plt.figure(figsize=(12, 8))
fig = sm.graphics.plot_ccpr_grid(beef_model, fig=fig)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(beef_model, 'Price', fig=fig)
```

```python
beef_model = smf.ols("Quantity ~ Price", data=beef).fit()
print(beef_model.summary())
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
dat = sm.datasets.get_rdataset("Guerry", "HistData").data
dat = dat.dropna()
results = smf.ols(r'Lottery ~ Literacy + np.log(Pop1831)', data=dat).fit()
print(results.summary())
```

```python
df = sm.datasets.get_rdataset("Guerry", "HistData").data
df = df[['Lottery', 'Literacy', 'Wealth', 'Region']].dropna()
df.head()
```

```python
mod = smf.ols(formula='Lottery ~ Literacy + Wealth + Region', data=df)
res = mod.fit()
print(res.summary())
```
